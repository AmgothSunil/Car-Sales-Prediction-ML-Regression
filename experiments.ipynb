```python
## lets drop carid and carname

df.drop(['car_ID','CarName'], axis=1,inplace=True)
```

```python
#Feature engineering
target = 'price'

categorical_columns = [column for column in df.columns if df[column].dtype=='object' and column != target]
categorical_columns
```

```python
numerical_columns = [column for column in df.columns if column not in categorical_columns and column != target]
numerical_columns
```

```python
len(numerical_columns)
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
X = df.drop('price', axis=1)
y = df['price']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=42)

print(len(X_train),len(X_test), len(y_train), len(y_test))
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_columns),
        ('cat', OneHotEncoder(handle_unknown='ignore', drop='first'), categorical_columns)
    ], remainder='passthrough'
)
```

```python
X_train_preprocessed = preprocessor.fit_transform(X_train)
X_test_preprocessed = preprocessor.transform(X_test)   # we don't do fit_transform here beacuse, there will be chance of data leakage
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
models = [LinearRegression(), SVR(), KNeighborsRegressor(), DecisionTreeRegressor(), RandomForestRegressor(),
            GradientBoostingRegressor(), XGBRegressor()]

for model in models:
    model.fit(X_train_preprocessed, y_train)

    y_pred = model.predict(X_test_preprocessed)

    print(f"model name : {model}")
    print(f'{model.__class__.__name__} params : {model.get_params()}')
    print("r2_score :", r2_score(y_test, y_pred))
    print("mse: ", mean_squared_error(y_test, y_pred))
    print("mae: ", mean_absolute_error(y_test, y_pred))
    print('\n')
    print(f'------------------------------')

```

```python
# the best performing model is random forest model

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_forest_params = {
    'n_estimators': [100, 200, 300],
    'criterion': ["squared_error", "absolute_error"],
    'max_features': ["sqrt", "log2", 0.5], 
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
}
```

```python
grid_search = RandomizedSearchCV(
    estimator=RandomForestRegressor(),
    param_distributions=random_forest_params,
    n_iter=50,
    scoring='f1',
    cv=5,
    n_jobs=-1,
    verbose=2,
    random_state=42
)
```

```python
random_search.fit(X_train_preprocessed, y_train)
```

```python
y_pred = random_search.predict(X_test_preprocessed)
print("r2_score :", r2_score(y_test, y_pred))
print("mse: ", mean_squared_error(y_test, y_pred))
print("mae: ", mean_absolute_error(y_test, y_pred))
print('\n')
print(f'------------------------------')

```

```python
random_search.best_params_
```

```python
model = RandomForestRegressor(n_estimators=100,min_samples_split=5,min_samples_leaf=2,max_features='log2',max_depth=None,criterion='absolute_error')

model.fit(X_train_preprocessed, y_train)

y_pred = model.predict(X_test_preprocessed)
print("r2_score :", r2_score(y_test, y_pred))
print("mse: ", mean_squared_error(y_test, y_pred))
print("mae: ", mean_absolute_error(y_test, y_pred))
print('\n')
print(f'------------------------------')
```
